```python
import json
import pandas as pd
import numpy as np
```

```python
data = json.load(open("sample-1.json"))
courses = [];
details = [];

for key, value in data.items():
    print("course: ", key)
    courses.append(key)

for course in courses:
    for key, value in data[course].items():
        print("details_for: ",key)
        details.append(value)
    
details
```

```python
pd.json_normalize(details)
```
